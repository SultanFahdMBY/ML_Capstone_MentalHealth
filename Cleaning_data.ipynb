### Cleaning Data

```python
df = df_1.drop(columns=['tweet_id'])
df = df_1[['content','sentiment']]
```

```python
df = df.rename(columns={'content': 'Text', 'sentiment': 'Mood'})
df
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df = df.query("Mood != 'surprise'")
df = df.query("Mood != 'boredom'")
df = df.query("Mood != 'empty'")

df['Mood'].value_counts()
```

```python
df
```

```python
df_2 = df_2.rename(columns={'Comment': 'Text', 'Emotion': 'Mood'})
df_2
```

```python
df_3 = df_3.rename(columns={'text': 'Text', 'label': 'Mood'})
df_3
```

```python
df_3[df_3["Mood"] == 5]
```

```python
#Drop category Mood = 5
df_3 = df_3.query("Mood != 5")
```

```python
#Rename coloumn dataframe 3
df_3['Mood'] = df_3['Mood'].astype('category')
df_3['Mood'] = df_3['Mood'].cat.rename_categories({0: 'sadness', 1: 'joy', 2: 'love', 3: 'anger', 4:'fear'})
```

```python
import string
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

df_total["Text"] = df_total['Text'].apply(remove_punctuations)
```

#### Random Data

```python
# Mengambil 5000 data setiap category mood
def random_sample(data, total, label, state=None):
    mood_data= data[data['Mood']==label]
    random_samplee = mood_data.sample(n=total, random_state=state)
    return random_samplee
```
